```python
# import modules
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
```

# Extract data from files

```python
# read data from file
f = open("data", 'r')
lines = f.readlines()
f.close()
at_aver = []
bst_aver = []
for line in lines:
    l = line.split()
    at_aver.append(int(l[0]))
    bst_aver.append(int(l[1]))
aver_data = np.zeros((len(at_aver),2), dtype=int)
for i in range(len(at_aver)):
    aver_data[i,0] = at_aver[i]
    aver_data[i,1] = bst_aver[i]
df = pd.DataFrame(aver_data)
df.columns = ['<at_i+1 - at_i>', '<bst>']
df.head()
```

```python
# read RR 1
f = open("RR_ALG+1", 'r')
lines = f.readlines()
f.close()
tat = []
wt = []
for line in lines:
    l = line.split()
    tat.append(float(l[1]))
    wt.append(float(l[0]))
tat = np.array(tat)
wt = np.array(wt)
df['<wt> (RR_1)'] = wt
df['<tat> (RR_1)'] = tat
df.head()
```

# build data frame with algorithm minimum wt

```python
df2 = df.copy()
df2 = df2.drop(['<at_i+1 - at_i>'], axis = 1)
df2 = df2.drop(['<bst>'], axis = 1)
df2 = df2.drop(['<tat> (RR_1)'], axis = 1)
df2 = df2.drop(['<tat> (RR_2)'], axis = 1)
df2 = df2.drop(['<tat> (RR_3)'], axis = 1)
df2 = df2.drop(['<tat> (RR_4)'], axis = 1)
df2 = df2.drop(['<tat> (RR_5)'], axis = 1)
minValuesdf = df2.idxmin(axis=1)
#
df3 = df.copy()
df3 = df3.drop(['<tat> (RR_1)'], axis = 1)
df3 = df3.drop(['<tat> (RR_2)'], axis = 1)
df3 = df3.drop(['<tat> (RR_3)'], axis = 1)
df3 = df3.drop(['<tat> (RR_4)'], axis = 1)
df3 = df3.drop(['<tat> (RR_5)'], axis = 1)
df3 = df3.drop(['<wt> (RR_1)'], axis = 1)
df3 = df3.drop(['<wt> (RR_2)'], axis = 1)
df3 = df3.drop(['<wt> (RR_3)'], axis = 1)
df3 = df3.drop(['<wt> (RR_4)'], axis = 1)
df3 = df3.drop(['<wt> (RR_5)'], axis = 1)
df3['ALGO'] = minValuesdf
```

# Find algorithm minimum turn around time

```python
df2 = df.copy()
df2 = df2.drop(['<at_i+1 - at_i>'], axis = 1)
df2 = df2.drop(['<bst>'], axis = 1)
df2 = df2.drop(['<wt> (RR_1)'], axis = 1)
df2 = df2.drop(['<wt> (RR_2)'], axis = 1)
df2 = df2.drop(['<wt> (RR_3)'], axis = 1)
df2 = df2.drop(['<wt> (RR_4)'], axis = 1)
df2 = df2.drop(['<wt> (RR_5)'], axis = 1)
minValuesdf = df2.idxmin(axis=1)
#
df4 = df.copy()
df4 = df4.drop(['<tat> (RR_1)'], axis = 1)
df4 = df4.drop(['<tat> (RR_2)'], axis = 1)
df4 = df4.drop(['<tat> (RR_3)'], axis = 1)
df4 = df4.drop(['<tat> (RR_4)'], axis = 1)
df4 = df4.drop(['<tat> (RR_5)'], axis = 1)
df4 = df4.drop(['<wt> (RR_1)'], axis = 1)
df4 = df4.drop(['<wt> (RR_2)'], axis = 1)
df4 = df4.drop(['<wt> (RR_3)'], axis = 1)
df4 = df4.drop(['<wt> (RR_4)'], axis = 1)
df4 = df4.drop(['<wt> (RR_5)'], axis = 1)
df4['ALGO'] = minValuesdf
```

```python
df4.head()
```

# Apply KNN to turn around time

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
X = df4.drop(['ALGO'], axis=1)
y = df4['ALGO']
knn = KNeighborsClassifier(n_neighbors=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=32)
# train the model
knn.fit(X_train, y_train)
s=knn.score(X_test, y_test)
print(s)
# predict response for test
y_pred = knn.predict(X_test)
y_pred
y_test
```
